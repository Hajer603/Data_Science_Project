# **Filling NaN**

```python
# First I'll change some type of data so that I can fill them with the write way
Hilal_df["Garage"] = pd.to_numeric(Hilal_df["Garage"], errors="coerce")
```

```python
#now I'll check if it changed
Hilal_df.info()
```

```python
#now I can start filling NaN
```

### **City**

**I choos eto fill it with the mode because ONLY 2 values are missing, and cities are
usually repeated in listings. Using the most common city avoids introducing noise, 
while "Unknown" keeps missing values identifiable if needed.**

```python
Hilal_df["City"].fillna(Hilal_df["City"].mode()[0], inplace=True)
```

```python
Hilal_df.isnull().sum()
```

### **Beds and Baths**

**I choose to fill them with Median because Beds/Baths are counts (e.g., 1, 2, 3, 4).
Median is robust to outliers. If a few luxury properties have 10+ bedrooms, they won’t distort the median.
Using mean could give fractional results (e.g., 2.4 bathrooms), which isn't realistic for counts.**

```python
Hilal_df["Beds"].fillna(Hilal_df["Beds"].median(), inplace=True)
Hilal_df["Baths"].fillna(Hilal_df["Baths"].median(), inplace=True)
```

```python
Hilal_df.isnull().sum()
```

### **Garage**

**I choose to fill these NaN with (0) because In real estate, a missing garage usually means no garage.
Filling with 0 is semantically meaningful and avoids falsely inflating the number of garages.
Alternative: median() if you suspect garages were omitted by mistake and most properties have garage**

```python
Hilal_df["Garage"].fillna(0, inplace=True) 
```

```python
Hilal_df.isnull().sum()
```

### **Area**

**I choose to fill theses data with Median because Area can have large variations (e.g., 80 m² to 1200 m²).
Median again avoids skew from large luxury properties.
Helps maintain central tendency and keeps distributions realistic.**

```python
Hilal_df["Area"].fillna(Hilal_df["Area"].median(), inplace=True)
```

```python
Hilal_df.isnull().sum()
```
