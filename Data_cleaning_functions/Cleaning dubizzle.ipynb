# **Filling NaN data**

### **Filling for prop_bathrooms and number of rooms**

```python
# I choose to fill this type of data with (Median) because it robust to outliers and good for room/bathroom counts
# To do this, I have first to convert the type of data from (Object) to (numeric)
# Convert to numeric (non-numeric values will become NaN automatically)
Dubizzle_df["prop_bathrooms"] = pd.to_numeric(Dubizzle_df["prop_bathrooms"], errors="coerce")
Dubizzle_df["number_of_rooms"] = pd.to_numeric(Dubizzle_df["number_of_rooms"], errors="coerce")
```

```python
#Know I'll check of the conversion oocoured
Dubizzle_df.info()
```

```python
#So now I can fill the missing values with the (Median)
Dubizzle_df["prop_bathrooms"].fillna(Dubizzle_df["prop_bathrooms"].median(), inplace=True)
Dubizzle_df["number_of_rooms"].fillna(Dubizzle_df["number_of_rooms"].median(), inplace=True)
```

```python
#Check if I filled all NaN
Dubizzle_df.isnull().sum()
```

```python
Dubizzle_df
```

```python
#I'll remove the units from the columns for (sale_price) and (prop_area)
#Ensure the column is string
Dubizzle_df["sale_price"] = Dubizzle_df["sale_price"].astype(str)
#Extract the numeric part
Dubizzle_df["sale_price"] = Dubizzle_df["sale_price"].str.extract(r'(\d+\.?\d*)')[0]

#Convert to float
Dubizzle_df["sale_price"] = Dubizzle_df["sale_price"].astype(float)

#Convert from hundreds to thousands, then to int
Dubizzle_df["sale_price"] = (Dubizzle_df["sale_price"] * 1000).astype(int)
```

```python
# Convert to string in case it's not
Dubizzle_df["prop_area"] = Dubizzle_df["prop_area"].astype(str)
#Extract the numeric part
Dubizzle_df["prop_area"] = Dubizzle_df["prop_area"].str.extract(r'(\d+\.?\d*)')[0]

#Convert to float
Dubizzle_df["prop_area"] = Dubizzle_df["prop_area"].astype(float)

#Convert from hundreds to thousands, then to int
Dubizzle_df["prop_area"] = (Dubizzle_df["prop_area"] * 1).astype(int)
```
